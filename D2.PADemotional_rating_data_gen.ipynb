Rate each discrete emotion in 3 dimension of PAD model: Pleasure, Arousal, Dominance. The table values are taken from the following papers

1.	Bradley, M. M., & Lang, P. J. (1999). Affective norms for English words (ANEW): Instruction manual and affective ratings (Vol. 30, No. 1, pp. 25-36). Technical report C-1, the center for research in psychophysiology, University of Florida.
2.	Warriner, A. B., Kuperman, V., & Brysbaert, M. (2013). Norms of valence, arousal, and dominance for 13,915 English lemmas. Behavior research methods, 45(4), 1191-1207.

```python
# Calculate pleasant value for each row

pleasant=emotion_weights_p.sum(axis=1)/no_of_emotion
arousal=emotion_weights_a.sum(axis=1)/no_of_emotion
dominance=emotion_weights_d.sum(axis=1)/no_of_emotion



time_frame.insert(loc=337, column='pleasant_score', value=pleasant)

time_frame.head()

```

```python
# Calculate arousal score for each row

time_frame.insert(loc=338, column='arousal_score', value=arousal)

time_frame.head()
```

```python
# Calculate dominance score for each row
time_frame.insert(loc=339, column='dominance_score', value=dominance)

time_frame.head()
```

```python
#calculate dominant emotion

#subtract 5 to scale the score between -4 to 4

p_score=time_frame['pleasant_score']-5
a_score=time_frame['arousal_score']-5
d_score=time_frame['dominance_score']-5



emo_frame=np.empty([len(time_frame.index), 3])

emo_frame[:,0]=p_score.abs()
emo_frame[:,1]=a_score.abs()
emo_frame[:,2]=d_score.abs()
emo_frame

max_v=np.argmax(emo_frame, axis=1)

print(max_v)

label=np.empty([len(time_frame.index)])

# Assign labels to the dominant emotion based on value and sign of each emotional score in 3 dimension

for i in range(len(max_v)):
    index=max_v[i]
    if index==0:
        if p_score.iloc[i]<0:
            label[i]=1 #discordant
        else:
            label[i]=2 #pleasant
    elif index==1:
        if a_score.iloc[i]<0:
            label[i]=3 #dissuade
        else:
            label[i]=4 #aroused
            
    elif index==2:
        if d_score.iloc[i]<0:
            label[i]=5 #submissive
        else:
            label[i]=6 #dominance
            
label
```
