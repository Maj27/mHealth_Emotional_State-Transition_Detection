Create minute-based temporal features to re-sample data later

```python
import pandas as pd
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

#load file created in step D2
path ='data/emotional_rating_18P_allF.csv' # use your path

df= pd.read_csv(path)
df.head()
```

```python
df.shape
```

```python
list(df)
```

```python
#clean data
df.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1, inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
sel_feat_df=df.drop(['pleasant_score','arousal_score', 'dominance_score', 'emotion'], axis=1)
sel_feat_df.shape
```

```python
# create more temporal features: minute of the day and minute of the hour
t_df=pd.DataFrame()
t_df['timestamp']=pd.to_datetime(sel_feat_df['timestamp'], unit='s')
df_time=pd.DataFrame()

df_time['minute_of_hour'] = t_df['timestamp'].dt.minute
df_time['minute_of_day'] =df_time['minute_of_hour']+sel_feat_df['hour_of_timestamp']*60

df_time
```

```python
sel_feat_df['minute_of_hour']= df_time['minute_of_hour']
sel_feat_df['minute_of_day']= df_time['minute_of_day']
sel_feat_df.head()
```

```python
sel_feat_df.drop('geohash', axis=1, inplace=True)
```

```python
sel_feat_df.shape
```

```python
sel_feat_df['pleasant_score']= df['pleasant_score']
sel_feat_df['arousal_score']= df['arousal_score']
sel_feat_df['dominance_score']= df['dominance_score']
sel_feat_df['emotion']= df['emotion']
#sel_feat_df.head()
```
