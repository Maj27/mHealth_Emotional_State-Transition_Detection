```python
diff_df=df.diff().abs()
    
```

```python
diff_df.head()
```

```python
# drop the first row since the change will be counted from the second row
diff_df.drop([0], axis=0, inplace=True)
diff_df.head()
```

```python
# calculate if there is any emotional transision from the previous window or not
# Assign 1 for transision and 0 for no-transition
def label_EmotionChange (row):
    if row['emotion']> 0:
        return 1
    else:
        return 0
```

```python
# apply the function to each row
diff_df['emotion_change'] = diff_df.apply (lambda row: label_EmotionChange (row), axis=1)
```

```python
#drop emotion variable since we will consider only emotional transition this time
diff_df.drop(['emotion'], axis=1, inplace=True)
```

```python
#save one file for each person

diff_df.to_csv('data/NonOverlap_w5_emoChange_SelFeat_data_p'+str(p)+'.csv', index=False)
```
